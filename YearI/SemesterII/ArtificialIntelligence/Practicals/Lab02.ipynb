### Implement the descriptive statistics for any numerical features using pandas libraries.

```python
from pandas import DataFrame
from pandas import read_csv

# Create DataFrame
dataset = read_csv("Dataset/cyber_crimes.csv")
df = DataFrame(dataset)
print(df)
```

```python
# Print Count of Dataset
print("\nCount of Dataset:")
counts = df['S. No'].count()
print(counts)
```

```python
# Print total crimes in each year
print("\nTotal crimes in each year:")
sum1 = df['2016'].sum()
print('2016: ', sum1)
sum2 = df['2017'].sum()
print('2017: ', sum2)
sum3 = df['2018'].sum()
print('2018: ', sum3)
```

```python
import matplotlib.pyplot as plt
plt.pie([sum1, sum2, sum3], labels=['2016', '2017', '2018'], colors=['red', 'blue', 'green'], autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# Print mean of crimes each year
print("\nMean of crimes each year:")
mean = df['2016'].mean()
print('2016: ', mean)
mean = df['2017'].mean()
print('2017: ', mean)
mean = df['2018'].mean()
print('2018: ', mean)
```

```python
# Print median of crimes each year
print("\nMedian of crimes each year:")
median = df['2016'].median()
print('2016: ', median)
median = df['2017'].median()
print('2017: ', median)
median = df['2018'].median()
print('2018: ', median)
```

```python
# Print mode of crimes each year
print("\nMode of crimes each year:")
mode = df['2016'].mode()
print('\n2016: ', mode)
mode = df['2017'].mode()
print('\n2017: ', mode)
mode = df['2018'].mode()
print('\n2018: ', mode)
```

```python
# Print maximum crimes each year
print("\nMaximum crimes each year:")
max = df['2016'].max()
print('\n2016: ', max, ',', DataFrame(df[df['2016'] == max])["State/UT"].values[0])
max = df['2017'].max()
print('\n2017: ', max, ',', DataFrame(df[df['2017'] == max])["State/UT"].values[0])
max = df['2018'].max()
print('\n2018: ', max, ',', DataFrame(df[df['2018'] == max])["State/UT"].values[0])
```

```python
# Print standard deviation of crimes each year
print("\nStandard deviation of crimes each year:")
std = df['2016'].std()
print('\n2016: ', std)
std = df['2017'].std()
print('\n2017: ', std)
std = df['2018'].std()
print('\n2018: ', std)
```
